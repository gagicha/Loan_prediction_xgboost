```python
import pandas as pd
from sklearn import preprocessing, metrics, model_selection, linear_model
from xgboost import XGBClassifier
```

```python
train=pd.read_csv("train_u6lujuX_CVtuZ9i.csv")
test=pd.read_csv("test_Y3wMUE5_7gLdaTN.csv")
```

```python
train.columns
```

```python
train=train.drop('Loan_ID',1)
test=test.drop('Loan_ID',1)
```

```python
train.shape
```

```python
train['Married'].isnull().sum()
```

```python
train.fillna(train.dropna().max(),inplace =True)
```

```python
test.fillna(test.dropna().max(),inplace =True)
```

```python
train.dtypes
```

```python
le= preprocessing.LabelEncoder()
for i in train.columns:
    if train[i].dtype== type(object):
        train[i]=le.fit_transform(train[i])
for i in test.columns:
    if test[i].dtype== type(object):
        test[i]=le.fit_transform(test[i])
        
```

```python
y= train['Loan_Status']
x=train.drop('Loan_Status',1)
```

```python
train_x, valid_x, train_y, valid_y= model_selection.train_test_split(x,y,test_size=0.2)
```

```python
model= XGBClassifier(max_depth=2,base_score=0.9, n_estimators=50)
```

```python
model.fit(train_x, train_y)
```

```python
pred=model.predict(valid_x)
```

```python
print(pred)
```

```python
# metrics.accuracy_score(valid_y, pred)
model.score(valid_x, valid_y)
```

```python
print(metrics.confusion_matrix(valid_y, pred))
```
